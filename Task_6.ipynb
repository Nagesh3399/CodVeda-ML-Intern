### K-means Clustering

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv(r"C:\Users\N.NAGESH\Desktop\CodVeda\Dataset\1) iris.csv")
```

```python
data.head()
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
model  = KMeans(n_clusters= 3)
model.fit(x_train)
```

```python
labels = model.labels_
centroid = model.cluster_centers_
```

```python
pca = PCA(n_components=2)
df_2D = pca.fit_transform(x_train)
centroids_2D = pca.transform(centroid)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_2D[:, 0], df_2D[:, 1], c=labels, cmap='viridis', alpha=0.6)
plt.scatter(centroids_2D[:, 0], centroids_2D[:, 1], c='red', marker='X', s=200, label='Centroids')

# Draw lines from each point to its centroid
for i in range(len(df_2D)):
    plt.plot([df_2D[i, 0], centroids_2D[labels[i], 0]], 
             [df_2D[i, 1], centroids_2D[labels[i], 1]], 
             'k-', alpha=0.3, linewidth=0.5)

plt.legend()
plt.title("K-Means Clustering (PCA-Reduced to 2D)")
plt.show()
```
